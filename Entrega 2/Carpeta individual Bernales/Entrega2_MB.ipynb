```python
import csv

# Abre el archivo CSV original y lee su contenido
with open('/content/MBERNALES_CICLONES.csv', 'r', encoding='utf-8') as archivo_entrada:
    lector_csv = csv.reader(archivo_entrada, delimiter=';')
    filas = list(lector_csv)

# Escribe el contenido en un nuevo archivo CSV con comas como delimitador
with open('archivo_modificado.csv', 'w', encoding='utf-8', newline='') as archivo_salida:
    escritor_csv = csv.writer(archivo_salida, delimiter=',')
    escritor_csv.writerows(filas)

print("Archivo modificado exitosamente.")
```
